# Xử lý dataset ECML

```
# @title Import thư viện panda
import pandas as pd
```

```
# @title Đọc dataset **ecml** và lưu vào Data Frame **d1** rồi xử lý
d1 = pd.read_csv("drive/MyDrive/Machine_Log_Analysis/Dataset/ecml/ecml.csv")

# @markdown - Giữ lại các cột type, method, uri, query và body
d1 = d1[['ns1:type','ns1:method','ns1:uri','ns1:query','ns1:body']]

# @markdown - Loại bỏ các hàng có type = NaN
d1 = d1.dropna(subset = ['ns1:type'])

# @markdown - Gộp cột query và cột body vào thành cột **result**
d1.loc[:, 'result'] = d1['ns1:query'].fillna('') + d1['ns1:body'].fillna('')

# @markdown - Giữ lại cột type, uri và result
d1 = d1.loc[:, ['ns1:type','ns1:uri','result']]

# @markdown - Loại bỏ tất cả các cột có result trống
d1.loc[d1['result']!='']

# @markdown - Đổi tên cột uri => path và cột result => query
d1.rename(columns = {'ns1:type':'type','ns1:uri':'path','result':'query'}, inplace = True)
```

# Tạo một DataFrame mới tên là df, sau đó gộp tất cả d1 và c1-c8 vào df để tạo final_dataset

```
# @title Gộp tất cả vào DataFrame df

df = pd.concat([d1,c1]).reset_index(drop = True)
df = pd.concat([df,c2]).reset_index(drop = True)
df = pd.concat([df,c3]).reset_index(drop = True)
df = pd.concat([df,c4]).reset_index(drop = True)
df = pd.concat([df,c5]).reset_index(drop = True)
df = pd.concat([df,c6]).reset_index(drop = True)
df = pd.concat([df,c7]).reset_index(drop = True)
df = pd.concat([df,c8]).reset_index(drop = True)
df
```

```
# @title Xem thông tin của df
df.info()
```

```
# @title Rút gọn tên rồi nhóm và đếm số lượng theo type

df.loc[df['type']=='OsCommanding','type'] = 'OS'
df.loc[df['type']=='LdapInjection','type'] = 'LDAPi'
df.loc[df['type']=='SqlInjection','type'] = 'SQLi'
df.loc[df['type']=='PathTransversal','type'] = 'PT'
df.loc[df['type']=='XPathInjection','type'] = 'XPath'
df.groupby(['type']).count()
```

```
# @title Tạo url từ path và query

# @markdown - Thao tác:
# @markdown - Những dòng có query != NaN, gộp cột path và query thành cột url, ngăn cách bởi một dấu ?, nếu path = NaN thì chuyển thành " " trước khi gộp
# @markdown - Những dòng có query == NaN, gán giá trị của query vào cột url, nếu path = NaN thì chuyển thành " " trước khi gán
# @markdown - Loại bỏ quequery và path, chỉ giữ lại url

df.loc[~df['query'].isnull(),'url'] = df.loc[~df['query'].isnull(),'path'].fillna('') + '?' + df.loc[~df['query'].isnull(),'query'].fillna('')
df.loc[df['query'].isnull(),'url'] = df.loc[df['query'].isnull(),'path'].fillna('')
df = df.loc[:, ['type','url']]
df
```

```
# @title Chuyển df thành stage.csv rồi tải xuống

from google.colab import files
df.to_csv('/content/drive/MyDrive/Machine_Log_Analysis/Dataset/stage.csv')
files.download('stage.csv')
```
